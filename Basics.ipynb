# Changes in Euro-Dollar Exchange Rates During the Coronavirus Pandemic

In this project, we will be analyzing the data of the Euro-USD currency exchange rate over the years and see the changes that have occurred as a result of the recent COVID-19 pandemic.

This dataset involves exchange rates from 1999 to 2021. We will focus on the years 2016-2019, before the pandemic, and 2020-2021, after the pandemic.

To preface this project, we will define some terms and events:
- The COVID-19 pandemic regards to the worldwide outbreak of the Coronavirus disease starting in 2020.

    - This induced quarantines and shut-downs for in-person activities and gatherings. 
    - This impacted many sectors of the economy, such as in-person work, vacationing, or technology, and caused a sudden growth in demand throughout the course of the event.
    
- Euros are considered the official currency of the European Union.

- The Euro-USD exchange rate refers to the rate at which the Euro will be exchanged for another currency, the US Dollar.

    - Example: 15 Euro = 15.25 USD. 15.25/15 = 1.02/1
        - 1.02 US Dollars are exchanged for every 1 Euro. 

## Summary of Results

We found that within the years of 2016-2019, two peaks in exchange rates occurred above 1.2 in the years 2016 and 2018, and declined starting from the peak in 2018 up to 2021.
Starting from 2020 to 2021, the exchange rate began rising and peaked at a value above 1.2 yet again after 2 years of decline since 2018.

## Cleaning the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

Analyzing the data, of the 41 columns and 40 rows, containing a combination of float64 and object data types, around 42% of the columns contain null values.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['US_dollar', 'Time']].copy()
exchange_rates['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

euro_to_dollar.info()
```

## Visualizing the Overall Evolution of Exchange Rates

```python
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.xlabel('Time (years)'); plt.ylabel('US_dollar');
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

```python
euro_to_dollar['rolling_mean']
```

The visualization shows that the recent years surrounding and within the pandemic shows a moderate decline going into 2020 from around 2018, and then a growth starting from 2020. 

## Visualizing How the Euro-Dollar Rate Changed During the Coronavirus Pandemic

```python
baseline = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2016
                   ) & (euro_to_dollar['Time'].dt.year <= 2021)]
pandemic = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >= 2020
                   ) & (euro_to_dollar.Time.dt.year <= 2021)]

# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

# Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(baseline['Time'],
        baseline['rolling_mean'],
        linewidth=1, color='#A6D785')

# Highlighting the 2020 period
ax.plot(pandemic['Time'],
        pandemic['rolling_mean'],
        linewidth=3, color='#e23d28')
    
# Adding a title and a subtitle    
plt.suptitle("Euro-USD rate peaked at 1.22 Post-Coronavirus",
        fontsize=10,
        weight='bold',
        style='italic',
        family='monospace')
plt.title("Euro-USD rates Pre and Post-Coronavirus Era (2016-2019, 2020-2021)", fontsize= 6)

# Add some transparency to the graph
ax.tick_params(axis='x', colors='grey', labelsize=10)
ax.tick_params(axis='y', colors='grey', labelsize=10)

ax.set_alpha(0.01)

ax.grid(alpha=0.5)

# Adding a signature
ax.text(0.9, 0.1, 'Â©Saira Erfan', transform=ax.transAxes,
        fontsize=10, color='gray', alpha=0.5, rotation='0')
ax.text(0.9, 0.05, 'Source: European Central Bank', transform=ax.transAxes,
        fontsize=10, color='gray', alpha=0.5,
        ha='center', va='center', rotation='0')

plt.show()
```

## Conclusion

Using data regarding European exchange rates over the years, we utilized data targeting exchange rates with the U.S. dollar to visualize the change in rate pre and post-coronavirus pandemic.

Looking at the graph, our results indicate that:
- From 2016-2019 (pre-coronavirus), two peaks in exchange rates were reached
    - Early 2016: 1.13
    - Early 2018: 1.25
- From 2020-2021 (post-coronavirus), one peak in exchange rates was reached
    - Start of 2021: 1.22

Knowing this, we can assume that due to the COVID-19 pandemic, exchange rates slowly started to increase to a value of 1.22, higher than the rate had been since 2018. 

One of the factors that could have brought this steep recovery was the increase in demand and reopening of many sectors of the economy after the initial quarantine and international shut-downs caused by the virus outbreak in 2020. With vaccines slowly gaining traction in prevalence and more in-person gatherings allowed in policy, sectors reopenening after being completely shut down, such as travel by plane or automobile, caused a more sudden and steep growth in the exchange rate between Europe and the U.S.
